```python
import random

puntos = 0
puntos_maquina = 0
lista = ["Piedra", "Papel", "Tijeras"]

while puntos <= 3 or puntos_maquina <= 3:

    jugador = input("Elige Piedra, papel o tijeras").capitalize()
    maquina = random.choice(lista)

    if jugador == "Piedra":
        print("Jugador ha elegido", jugador)
        print("La máquina ha elegido", maquina)

        if maquina == jugador:
            puntos += 0
            puntos_maquina += 0
            print(f"Empate. Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

        elif maquina == "Papel" and jugador == "Piedra":
            puntos += 0
            puntos_maquina += 1
            print(f"!Qué pena¡ Sigue jugando. Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

        elif maquina == "Tijeras" and jugador == "Piedra":
            puntos += 1
            puntos_maquina += 0
            print(f"¡Felididades has ganado! Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

    elif jugador == "Papel":
        print("Jugador ha elegido", jugador) 
        print("La máquina ha elegido", maquina) 

        if maquina == jugador:
            puntos += 0
            puntos_maquina += 0
            print(f"Empate. Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

        elif maquina == "Piedra" and jugador == "Papel":
            puntos += 1
            puntos_maquina += 0
            print(f"¡Felididades has ganado! Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

        elif maquina == "Tijeras" and jugador == "Papel":
            puntos += 0
            puntos_maquina += 1
            print(f"!Qué pena¡ Sigue jugando. Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

    elif  jugador == "Tijeras":
        print("Jugador ha elegido", jugador) 
        print("La máquina ha elegido", maquina)

        if maquina == jugador:
            puntos += 0
            puntos_maquina += 0
            print(f"Empate. Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

        elif maquina == "Papel" and jugador == "Tijeras":
            puntos += 1
            puntos_maquina += 0
            print(f"¡Felididades has ganado! Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

        elif maquina == "Piedra" and jugador == "Tijeras":
            puntos += 0
            puntos_maquina += 1
            print(f"!Qué pena¡ Sigue jugando. Vais jugador {puntos} y máquina {puntos_maquina}")
            print("\n")

    else:
        print("No has elegido un valor correcto")

    if puntos == 3 or puntos_maquina == 3:
        print(f"El total de puntos es: {jugador} {puntos} / Máquina {puntos_maquina}")
        if puntos == 3:
            print("¡Has ganado")
            break
        else:
            print("¡Has perdido")
            break
   
Gato = "(\__/)""\n"" (>’.’<)""\n"" («)_(«)"
print("¡Gracias por jugar!""\n",Gato)
```
